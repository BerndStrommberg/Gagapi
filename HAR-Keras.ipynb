```python
df = pd.read_csv('tidy_dataset_training.csv')
```

```python
df.head()
```

Input shape ist (rows, columns)

```python
df.columns[20]
```

```python
x_train = df[df.columns[1:21]].values
```

```python
x_train.shape
```

```python
x_train[0][19]
```

```python
y_train_preset = df[df.columns[35]].values
```

```python
y_train_preset
```

```python
y_train_final = np.zeros(shape=(y_train_preset.shape[0] ,6))
```

```python
y_train_final[1]
```

```python
for item in range(0, y_train_preset.shape[0]):
    y_train_final[item][int(y_train_preset[item])-1] = 0.99
    print(y_train_final[item])
    print(y_train_preset[item])
```

### Test Data

### Netz programmieren
Input ist jetzt vorbereitet

```python
model = Sequential()
model.add(Dense(20, activation='relu', input_dim=(20)))
model.add(Dropout(0.2))
model.add(Dense(20, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(6, activation='softmax'))

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

```

```python
model.fit(x_train, y_train_final, epochs=1000, batch_size=200)
```

### Testing

```python
test_data = pd.read_csv("tidy_dataset_test.csv")
```

```python
x_test = test_data[test_data.columns[1:21]].values
```

```python
x_test.shape
```

```python
y_test = test_data[test_data.columns[35]].values
```

```python
y_test_final = np.zeros(shape=(y_test.shape[0] ,6))
```

```python
for item in range(0, y_test.shape[0]):
    y_test_final[item][int(y_test[item])-1] = 0.99
  
```

```python
classes = model.predict(x_test, batch_size=128)
```

```python
model.predict
```

```python
classes.shape
```

```python
score = model.evaluate(x_test, y_test_final, batch_size=128)
```
