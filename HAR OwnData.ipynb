```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
import numpy as np
import pandas as pd
```

```python
moving = pd.read_csv("Server/TestData2/TestDatalabeled/walking.csv")
```

```python
staying = pd.read_csv("Server/TestData2/TestDatalabeled/staying.csv")
```

```python
sitting = pd.read_csv("Server/TestData2/TestDatalabeled/sitting.csv")
```

```python
all_movement = pd.concat([moving,staying,sitting])
```

```python
all_movement = all_movement.sample(frac=1).reset_index(drop=True)
```

```python
all_movement = all_movement.drop('Unnamed: 0', axis=1)
```

```python
all_movement.shape
```

```python
train_data = all_movement[0:10000]
```

```python
train_data.to_csv("Server/TestData2/TestDatalabeled/train.csv")
```

```python
test_data = all_movement[10001:]
```

```python
test_data.to_csv("Server/TestData2/TestDatalabeled/test.csv")
```

Konkrete Trainingsdaten

```python
x_train = train_data[train_data.columns[0:20]].values
```

```python
y_train = train_data[train_data.columns[20]].values
```

```python
x_train.shape
```

```python
x_test = test_data[test_data.columns[0:20]].values
```

```python
x_test.shape
```

```python
y_test = test_data[test_data.columns[20]].values
```

```python
y_test_f = np.zeros(shape=(y_test.shape[0], 3))
```

```python
for item in range(0, y_test.shape[0]):
    y_test_f[item][int(y_test[item])-1] = 0.99
    print(y_test_f[item])
    print(y_test[item])
```

```python
y_train_f = np.zeros(shape=(y_train.shape[0], 3))
```

```python
for item in range(0, y_train.shape[0]):
    y_train_f[item][int(y_train[item])-1] = 0.99
    print(y_train_f[item])
    print(y_train[item])
```

```python
model = Sequential()
model.add(Dense(20, activation='relu', input_dim=(20)))
model.add(Dropout(0.2))
model.add(Dense(30, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(20, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(3, activation='softmax'))

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train_f, epochs=700, batch_size=100)
```

```python
score = model.evaluate(x_test, y_test_f, batch_size=128)
```
