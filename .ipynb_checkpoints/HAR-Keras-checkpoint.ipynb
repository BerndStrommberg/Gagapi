## Shuffeln um bessere Ergebnisse zu erzielen

```python
tidy = pd.read_csv('tidy_dataset.csv')
```

```python
tidy = tidy.sample(frac=1).reset_index(drop=True)
```

```python
tidy.shape
```

```python
labelsDict = {
    'WALKING': 1,
    'WALKING_UPSTAIRS': 2,
    'WALKING_DOWNSTAIRS': 3,
    'SITTING': 4,
    'STANDING': 5,
    'LAYING': 6
}
```

```python
tidy.to_csv(path_or_buf="tidy_dataset_all.csv")
```

```python
df = pd.read_csv('tidy_dataset_training.csv')
```

```python
df.head()
```

Input shape ist (rows, columns)

```python
x_train = df[df.columns[1:33]].values
```

```python
y_train_preset = df[df.columns[35]].values
```

```python
y_train_preset
```

```python
y_train_final = np.zeros(shape=(y_train_preset.shape[0] ,6))
```

```python
y_train_final[1]
```

```python
for item in range(0, y_train_preset.shape[0]):
    y_train_final[item][int(y_train_preset[item])-1] = 0.99
    print(y_train_final[item])
    print(y_train_preset[item])
```

### Test Data

### Netz programmieren
Input ist jetzt vorbereitet

```python
model = Sequential()
model.add(Dense(20, activation='relu', input_dim=32))
model.add(Dense(10, activation='relu'))
model.add(Dense(6, activation='softmax'))

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

```

```python
model.fit(x_train, y_train_final, epochs=20, batch_size=100)
```

### Testing

```python
test_data = pd.read_csv("tidy_dataset_test.csv")
```

```python
x_test = test_data[test_data.columns[1:33]].values
```

```python
y_test = test_data[test_data.columns[35]].values
```

```python
y_test_final = np.zeros(shape=(y_test.shape[0] ,6))
```

```python
for item in range(0, y_test.shape[0]):
    y_test_final[item][int(y_test[item])-1] = 0.99
    print(y_test_final[item])
    print(y_test[item])
```

```python
model.evaluate(x_test, y_test_final, batch_size=128)
```

```python
classes = model.predict(x_test, batch_size=128)
```

```python
classes.shape
```

```python
predictions = np.zeros(shape=(classes.shape[0], 1))
for item in range(0, classes.shape[0]):
    predictions[item] = np.argmax(classes[item]) + 1
    
```

```python
for item in range(0, y_test.shape[0]):
    print("Expected: " + str(y_test[item]) + " Predicted: " + str(predictions[item]))
```
