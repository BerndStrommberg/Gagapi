### Creating the neural network

```python
class neuralNetwork:
    #schon angepasst
    def __init__(self, inputnodes, hiddennodes1, hiddennodes2, outputnodes, learningrate):
        
        self.inodes = inputnodes
        self.hnodes1 = hiddennodes1
        self.hnodes2 = hiddennodes2
        self.onodes = outputnodes
        self.lr = learningrate
        
        #Gewichte für das Netz
        self.wih = numpy.random.normal(0.0, pow(self.hnodes1, -0.5), (self.hnodes1, self.inodes))
        self.whh = numpy.random.normal(0.0, pow(self.hnodes2, -0.5), (self.hnodes2, self.hnodes1))
        self.who = numpy.random.normal(0.0, pow(self.onodes, -0.5), (self.onodes, self.hnodes2))
        
        self.activation_function = lambda x: scipy.special.expit(x)
        pass
    
    def train(self, input_list, target_list):
        # convert input lists into 2d array
        inputs = numpy.array(input_list, ndmin=2).T
        targets = numpy.array(target_list, ndmin=2).T
        
        #calculate signals into hidden layer1
        hidden1_inputs = numpy.dot(self.wih, inputs)
        
        #calculate output from hidden layer1
        hidden1_outputs = self.activation_function(hidden1_inputs)
        
        #calculate signals into hidden layer 2
        hidden2_inputs = numpy.dot(self.whh, hidden1_outputs)
        
        #calculate outputs from hidden layer 2
        hidden2_outputs = self.activation_function(hidden2_inputs)
        
        #caclulate inputs in final layer
        final_inputs = numpy.dot(self.who, hidden2_outputs)
        
        #calculate outputs from final layer
        final_outputs = self.activation_function(final_inputs)
        
        #output layer error ( target-actual)
        output_errors = targets - final_outputs
        
        #hidden layer2 error is output_errors, split by weihgts recombined ar hidden nodes
        hidden2_errors = numpy.dot(self.who.T, output_errors)
        
        hidden1_errors = numpy.dot(self.whh.T, hidden2_errors)
        
        
        #update the wheigts for the links between the hidden and output layers
        self.who += self.lr * numpy.dot((output_errors * final_outputs * (1.0 - final_outputs)), numpy.transpose(hidden2_outputs))
        
        #update the wheigts for the links between hidden and hidden layer
        self.whh += self.lr * numpy.dot((hidden2_errors * hidden2_outputs * (1.0 - hidden2_outputs)), numpy.transpose(hidden1_outputs))
        
        #update the wheigts for the links between hidden and input layers 
        self.wih += self.lr * numpy.dot((hidden1_errors * hidden1_outputs * (1.0 - hidden1_outputs)), numpy.transpose(inputs))
        pass
    
    
    #Schon angepasst
    def query(self, input_list):
        #input array in 2d array
        inputs = numpy.array(input_list).T
        
        #signals in hidden layer 1
        hidden1_inputs = numpy.dot(self.wih, inputs)
        
        #emerging from hidden layer 1
        hidden1_outputs = self.activation_function(hidden1_inputs)
        
        #signals in hidden layer 2
        hidden2_inputs = numpy.dot(self.whh, hidden1_outputs)
        
        #emerging from hidden layer 2
        hidden2_outputs = self.activation_function(hidden2_inputs)
        
        
        #signals into final output layer
        final_inputs = numpy.dot(self.who, hidden2_outputs)
        
        #calculate the signals emerging from output layer
        final_outputs = self.activation_function(final_inputs)
        
        return final_outputs
        
        pass
```

```python
input_nodes = 32
hidden1_nodes = 20
hidden2_nodes = 10
output_nodes = 6

learning_rate = 0.3

n = neuralNetwork(input_nodes, hidden1_nodes, hidden2_nodes, output_nodes, learning_rate)

```

```python
print(n.wih.shape) 
print(n.inputs_list.shape)
```

```python
currentValue = 0
scaledValue = 0
label = 0
labels = {
    "WALKING":1,
    "WALKING_UPSTAIRS":2,
    "WALKING_DOWNSTAIRS":3,
    "SITTING":4,
    "STANDING":5,
    "LAYING":6
}
for row in range(1, df.shape[0]):
    #List with the inputs from one row
    inputsList = [] 
    
    for column in df.columns[0:32]:
        currentValue = df.loc[row, column]
        scaledValue = (currentValue + 1)/2
        inputsList.append(scaledValue)

    #putting input to an numpy array
    inputs = inputsList
    # label ändern (das muss noch schöner)
    label = labels[df.loc[row,df.columns[34]]]
    
    #creating the expected target
    targets = numpy.zeros(output_nodes) + 0.01
    targets[label-1] = 0.99
    
    print("Actual label: " + str(label))
    print("targets max value index: " + str(numpy.argmax(targets)+1))
    
    n.train(inputs, targets)
    
    
    
    
```

```python
def testNetwork(row):
    currentValue = 0
    scaledValue = 0
    inputsList = [] 

    for column in df.columns[0:32]:
        currentValue = df.loc[row, column]
        scaledValue = (currentValue + 1)/2
        inputsList.append(scaledValue)
        
    inputs = inputsList
    output = n.query(inputs)
    print("Actual label: " + str(labels[df.loc[row,df.columns[34]]]))
    print("Network output: " + str(numpy.argmax(output) + 1))
    print("Raw output:\n")
    print(output)
        
```
