```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('tidy_dataset_training.csv')
```

```python
labelsDict = {
    'WALKING': 1,
    'WALKING_UPSTAIRS': 2,
    'WALKING_DOWNSTAIRS': 3,
    'SITTING': 4,
    'STANDING': 5,
    'LAYING': 6
}
```

```python
x_train = df[df.columns[1:7]].values
```

```python
x_train.shape
```

```python
y_train_preset = df[df.columns[35]].values
```

```python
y_train = np.zeros(shape=(y_train_preset.shape[0],6))
```

```python
y_train[1]
```

```python
for i, activity in enumerate(y_train_preset):
    y_train[i][activity-1] = 0.99
    print(y_train[i])
    print(y_train_preset[i])
```

```python
model = Sequential()
model.add(Dense(20, activation='relu', input_dim=(6)))
model.add(Dropout(0.1))
model.add(Dense(20, activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(6, activation='softmax'))

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

```

```python
test_data = pd.read_csv("tidy_dataset_test.csv")
x_test = test_data[test_data.columns[1:7]].values
y_test = test_data[test_data.columns[35]].values
y_test_final = np.zeros(shape=(y_test.shape[0] ,6))
for item in range(0, y_test.shape[0]):
    y_test_final[item][int(y_test[item])-1] = 0.99
  
```

```python
modelAccuracy = {}
def fitAndEvaluateModel(epochs,batch_size):
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    model.fit(x_train, y_train, epochs=epochs, batch_size=batch_size)
    score = model.evaluate(x_test, y_test_final, batch_size=batch_size)
    
    modelAccuracy["Epochen: " + str(epochs) + "| Batchsize: " + str(batch_size)] = score
```

```python
modelAccuracy
```

```python
for i in range(10, 410, 100):
    for j in range(10, 110, 60):
        fitAndEvaluateModel(i, j)
```
